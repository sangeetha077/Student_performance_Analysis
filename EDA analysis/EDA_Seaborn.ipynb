## Data Visualization

```python
import pandas as pd
```

```python
student_data=pd.read_csv("../data/cleaned_student_data.csv")
```

```python
student_data.head()
```

```python
student_data['total_score'] = (student_data['math_score'] + student_data['reading_score'] + student_data['writing_score']) / 3
```

```python
student_data.head()
```

### Average Total Score by Gender

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 5))
sns.lineplot(data=student_data, x='gender', y='total_score', marker='o', linewidth=2, color='skyblue')

plt.title('Average Total Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Total Score')
plt.ylim(0, 100)
plt.grid(True)
plt.tight_layout()
plt.show()
```

Female students have a slightly higher overall academic performance compared to male students in this dataset.

## Average Score by Lunch Type

```python
student_data.groupby('lunch')[['math_score', 'reading_score', 'writing_score']].mean().plot(kind='bar', figsize=(10, 6))
plt.title('Average Scores by Lunch Type')
plt.ylabel('Average Score')
plt.show()
```

### Average Scores by Lunch Type

```python
student_data.groupby('test_preparation_course')[['math_score', 'reading_score', 'writing_score']].mean().plot(kind='bar', figsize=(10, 6))
plt.title('Average Scores by Lunch Type')
plt.ylabel('Average Score')
plt.show()
```

### Test Preparation Course Participation by Gender

```python
sns.countplot(x='gender', hue='test_preparation_course', data=student_data)
plt.title('Test Preparation Course Participation by Gender')
plt.show()
```

### Score Distribution by Gender for Each Subject

```python
df_melted = student_data.melt(id_vars='gender', value_vars=['math_score', 'reading_score', 'writing_score'],
                    var_name='Subject', value_name='Score')

plt.figure(figsize=(12,6))
sns.barplot(x='Subject', y='Score', hue='gender', data=df_melted)
plt.title('Score Distribution by Gender for Each Subject')
plt.show()
```
